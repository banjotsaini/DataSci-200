````python
# Q6-2 Grading Tag:

scores = { "a": 1, "c": 3, "b": 3, "e": 1,  "d": 2, "g": 2,
            "f": 4, "i": 1, "h": 4, "k": 5,  "j": 8, "m": 3,
            "l": 1, "o": 1, "n": 1, "q": 10, "p": 3, "s": 1,
            "r": 1, "u": 1, "t": 1, "w": 4,  "v": 4, "y": 4,
            "x": 8, "z": 10 }

def char_count(s: str) -> dict[str, int]:
    """
    Return a dict mapping each character in s to its count
    """
    counts: dict[str, int] = {}
    for c in s:
        counts[c] = counts.get(c, 0) + 1
    return counts



def score_word(word: str) -> int:
    """
    Calculate the score for a word using the provided scores dictionary
    """

    total = 0
    for letter in word.lower():
        total += scores.get(letter, 0)
    return total
    raise NotImplementedError(
        "Implement this function to score a word based on the scores dictionary."
    )


def validate_rack(rack: str) -> bool:
    """
    Return True only if wildcards < 2 and word is 1-7 characters long and contains only letters
    """
    
    if rack is None:
        return False
    
    if not (2 <= len(rack) <= 7):
        return False
    
    wildcard_count = 0
    for c in rack:
        if c == "*":
            wildcard_count += 1
            if wildcard_count > 2:
                return False
        elif not ('A' <= c <= 'Z' or 'a' < c < "Z"):
            return False
    return True
    raise NotImplementedError(
        "Implement this function to validate the rack against the rules in the description."
    )


def run_scrabble(rack: str, words: list[str]) -> dict[str, int]:
    """
    Given an input rack and the sowpods.txt words list, return a dictionary mapping an uppercase word to it highest
    possible scrabble score using the rack.
    """

    rack = rack.lower()

    rack_counter = char_count(rack)
    wildcard_total = rack_counter.get('*', 0)
    if '*' in rack_counter:
        del rack_counter['*']

    res: dict[str, int] = {}

    for word in words:
        word = word.strip().lower()
        if len(word) > len(rack):
            continue

        word_counter = char_count(word)
        need = 0

        for letter, cnt in word_counter.items():
            available = rack_counter.get(letter, 0)
            if cnt > available:
                need += cnt - available
            if need > wildcard_total:
                break
        if need > wildcard_total:
            continue

        tile_counter = rack_counter.copy()
        wild = wildcard_total
        score = 0

        for letter in sorted(word, key=lambda l: scores[l], reverse=True):
            if tile_counter.get(letter, 0) > 0:
                score += scores[letter]
                tile_counter[letter] -= 1
            else:
                wild -= 1
            res[word.upper()] = score
    return res


    raise NotImplementedError(
        "Implement this function to find all valid words and their scores based on the rack provided."
    )


if __name__ == "__main__":
    # The condition above checks if this script is being run directly or being imported. More info:
    # https://docs.python.org/3/library/__main__.html
    # https://realpython.com/if-name-main-python/#in-short-it-allows-you-to-execute-code-when-the-file-runs-as-a-script-but-not-when-its-imported-as-a-module

    rack = input("Enter your Scrabble rack")  # feel free to use input() to get the rack from the user.
    # TODO: Call the functions above to complete the scrabble solver
    if not validate_rack(rack):
        print("Invalid rack, 2-7 chars, Limit <= 2 '*' wildcards")
    try:
        with open(SOWPODS_PATH) as f:
            word_list = f.readlines()
    except FileNotFoundError:
        print("Error: sowpods.txt not found")

    valid_words = run_scrabble(rack, word_list)
    for w, s in valid_words.items():
        print(f"{w}: {s}")

````
